```python
sns.set_style('whitegrid')
# Load the Tips dataset

tips = sns.load_dataset("tips")




# Create a histogram of the total bill amounts

sns.histplot(data=tips, x="total_bill")
plt.show()
```

Any visualization can fall into one of the three categories.


*   Univariate - x only (contains one axis of info) 
*   Bivariate -x and y (contains two axis of info)

*  Trivariate - x, y, z (contains three axis of info)

```python
tips.head()
```

**Scatter Plot**

Can be used to show the relationship between two numeric variables

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(x='total_bill', y='tip', data=tips)
ax.set_title('Relationship between tip vs total_bill', pad=10)
plt.show()
```

```python
# Add customization
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
sns.scatterplot(x='total_bill', y='tip', hue='sex', size='size', sizes=(50, 200), data=tips )

# add labels and title
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship between Total Bill and Tip')
plt.show()
```

**Line Plot**

can be used to visualize the trends in data over time or other continuous variables

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
plt.subplots(1, 1, figsize=(10, 5))
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.show()
```
